```python
import tqdm
import numpy as np
import matplotlib.pyplot as plt
from mpmath import *
```

```python
values = np.linspace((0.5, -10), (0.5, 10), 10000)
outputs = []
for value in values:
    output = zeta(value[0], value[1]*1j)
    real, imag = output.real, output.imag
    outputs.append((real, imag))
outputs = np.array(outputs)
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(outputs[:, 0], outputs[:, 1], s=1)
```
