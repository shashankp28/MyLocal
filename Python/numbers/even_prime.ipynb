```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
```

```python
def is_prime(n):
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

def getPrimes(n):
    prime = [True for i in range(n + 1)]
    p = 2
    while (p * p <= n):
        if (prime[p] == True):
            for i in range(p * 2, n + 1, p):
                prime[i] = False
        p += 1
    prime[0] = False
    prime[1] = False
    r = []
    for p in range(n + 1):
        if prime[p]:
            r.append(p)
    return r
```

```python
N = 1000000
primes = set(getPrimes(N))

def checkIsSum(primes, target):
    ways = 0
    for prime in primes:
        if target - prime in primes:
            ways += 1
    return ways

total, correct = 0, 0
ways = []
for i in tqdm(range(4, N+1, 2)):
    total += 1
    way = checkIsSum(primes, i)
    if way == 0:
        print(i)
        break
    ways += [way]
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(list(range(4, N+1, 2)), ways, c='r', s=0.01)
```
