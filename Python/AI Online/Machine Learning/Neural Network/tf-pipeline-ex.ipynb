```python
import tensorflow as tf
```

```python
dataset = tf.data.Dataset.list_files("../input/movie-review/reviews/*/*", shuffle=False)
```

```python
for data in dataset.as_numpy_iterator():
    print(data)
```

```python
def read_file(file_path):
    text = tf.io.read_file(file_path)
    text = tf.strings.regex_replace(text, "<br />", "")
    text = tf.strings.regex_replace(text, "\n", "")
    return text
```

```python
for path in dataset.take(1):
    print(read_file(path))
```

```python
import os
def get_label(file_path):
    splits = tf.strings.split(file_path, os.path.sep)
    return splits[-2]
```

```python
def process_txt(file_path):
    return read_file(file_path), get_label(file_path)
```

```python
def check_empty(text, label):
    return tf.strings.length(text)!=0
```

```python
dataset = dataset.map(process_txt).filter(check_empty).shuffle(2)
```

```python
for text, label in dataset.as_numpy_iterator():
    print(text, label)
    print()
```
