```python
import gensim
import pandas as pd
```

```python
df = pd.read_json("../input/sports-review/Sports_and_Outdoors_5.json", lines=True)
df.head()
```

```python
df.shape
df.reviewText[0]
```

```python
reviews = df.reviewText.apply(gensim.utils.simple_preprocess)
reviews[:5]
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4
)
```

```python
model.build_vocab(reviews, progress_per=1000)
```

```python
model.train(reviews, total_examples=model.corpus_count, epochs=15)
```

```python
model.save("w2v_ex.model")
```

```python
model.wv.most_similar("awful")
```

```python
model.wv.similarity('good', 'great')
```

```python
model.wv.similarity('slow', 'steady')
```
