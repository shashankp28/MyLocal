```python
import gensim
import pandas as pd
```

```python
df = pd.read_json("../input/amazon-reviews/Cell_Phones_and_Accessories_5.json", lines=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.reviewText[0]
```

```python
review_text = df.reviewText.apply(gensim.utils.simple_preprocess)
```

```python
review_text[:5]
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4
)
```

```python
model.build_vocab(review_text, progress_per=1000)
```

```python
model.train(review_text, total_examples=model.corpus_count, epochs=5)
```

```python
model.save("w2v.model")
```

```python
model.wv.most_similar("good")
```

```python
model.wv.similarity("good", "good")
```
